```python
import numpy as np
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
jsn_path = Path().resolve() / "Logs" / "svc.jsonl"
data = pd.read_json(jsn_path , lines = True)
```

```python
params = pd.json_normalize(data['params'])
```

```python
df = pd.concat([data[["target"]], params], axis = 1)
df.target = df.target*-1
```

```python
df1 = df.query("target < -0.87")
```

```python
px.scatter_3d(df1, x = "C", y = "gamma", z = "target", color = "target")
```

```python
px.scatter(df1, x="C", y="target", color="target")
```
